```python
import pandas as pd
import numpy as np
df = pd.read_csv("./bank.csv", sep=";")
```

```python
df.head()
```

```python
df.tail()
```

```python
def replace_status(val):
    if val == "single":
        return 0
    else:
        return 1
df["marital"] = df["marital"].apply(replace_status,1)
```

```python
df.head()
```

```python
df["housing"] = df["housing"].map({
    "no": 0,
    "yes": 1
}.get)
df.head()
```

```python
df["loan"] = df["loan"].replace({
    "no": 0,
    "yes": 1
})
```

```python
df.head()
```

```python
df["job"].unique()
```

```python
df["job"].replace({
    'unknown': np.nan,
    'management': 0,
    'technician': 1,
    'entrepreneur': 2,
    'blue-collar': 3,
    'retired': 4,
    'admin': 5,
    'services': 6,
    'self-employed': 7,
    'unemployed': 8,
    'housemaid': 9,
    'student': 10,
    'admin.': 11
}, inplace = True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["education"].unique()
```

```python
df["education"].replace({
    'primary': 0,
    'secondary': 1,
    'tertiary': 2,
    'unknown': np.nan
}, inplace = True)
```

```python
df.head()
```

```python
df['default'].unique()
```

```python
df['default'].replace({
    'no': 0,
    'yes': 1
}, inplace = True)
```

```python
df.head()
```

Applying Rescaling or Min-Max Normalization Technique on balance column

lambda in below statement means applying this on every row

```python
df["balance"] = df["balance"].apply(
lambda v: (v-df["balance"].min()) / (df["balance"].max() - df["balance"].min()))
```

```python
df.head()
```

```python
 df["contact"].unique()
```

```python
df["contact"].replace({
    "cellular": 1,
    "telephone": 0,
    "unknown": np.nan
}, inplace = True)
```

```python
df.head()
```

```python
df["month"].unique()
```

```python
df["month"].replace({
    "jan": 1,
    "feb": 2,
    "mar": 3,
    "apr": 4,
    "may": 5,
    "jun": 6,
    "jul": 7,
    "aug": 8,
    "sep": 9,
    "oct": 10,
    "nov": 11,
    "dec": 12
}, inplace = True)
```

```python
df.head()
```

```python
df.poutcome = df.poutcome.map({
    "unknown": np.nan,
    "failure": 0,
    "success": 2,
    "other": 1
})
```

```python
df.head()
```

```python
df.y.replace({
    "no": 0,
    "yes": 1
}, inplace = True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.y.unique()
```

```python
df["duration"] = df["duration"].apply(
lambda v: (v-df["duration"].min()) / (df["duration"].max() - df["duration"].min()))
```

```python
df.head()
```

```python
df.describe()
```

```python
df.to_csv("./bank.csv", index = False)
```
